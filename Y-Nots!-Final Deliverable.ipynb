## Null / NaN values
Data in our real world are rarely clean and homogeneous. Typically, they tend to be incomplete, noisy, and inconsistent. It is an important task for a Data scientist to prepossess the data by filling the missing values. This is very important to be handled since they could lead to wrong prediction or classification.

```python
# Displaying the count of Null values in each column
original_data.isnull().sum()
```

```python
# Creating multiple dataframes based upon Country column
data1=original_data[['Country','GDP']]
data2=original_data[['Country','Population']]
data3=original_data[['Country','Schooling']]
data4=original_data[['Country','Alcohol']]
data5=original_data[['Country','BMI']]
data6=original_data[['Country','Total_expenditure']]
data7=original_data[['Country','thinness__1-19_years']]
data8=original_data[['Country','Adult_Mortality']]
data9=original_data[['Country','Polio']]
data10=original_data[['Country','Diphtheria']]
data11=original_data[['Country','Hepatitis_B']]
```

```python
# Grouping and displaying the dataframe with the mean values of the specific country column
grouped1 = data1.groupby('Country').mean()
grouped2 = data2.groupby('Country').mean()

print(grouped1)
print(grouped2)
```

```python
# Grouping by country and replacing NULL values with country specific mean values

data1["GDP"] = data1.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data2["Population"] = data2.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data3["Schooling"] = data3.groupby("Country").transform(lambda x: x.fillna(0))
data4["Alcohol"] = data4.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data5["BMI"] = data5.groupby("Country").transform(lambda x: x.fillna(0))
data6["Total expenditure"] = data6.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data7["thinness__1-19_years"] = data7.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data8["Adult Mortality"] = data8.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data9["Polio"] = data9.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data10["Diphtheria"] = data10.groupby("Country").transform(lambda x: x.fillna(x.mean()))
data11["Hepatitis_B"] = data10.groupby("Country").transform(lambda x: x.fillna(x.mean()))

print(data1.loc[data1['Country'] == 'Somalia'])
print(data2.loc[data2['Country'] == 'Iraq'])
print(data3.loc[data3['Country'] == 'Cook Islands'])
print(data4.loc[data4['Country'] == 'Iraq'])
print(data5.loc[data5['Country'] == 'South Sudan'])
print(data6.loc[data6['Country'] == 'Algeria'])
print(data7.loc[data7['Country'] == 'Algeria'])
print(data8.loc[data8['Country'] == 'Iraq'])
print(data9.loc[data9['Country'] == 'Iraq'])
print(data10.loc[data10['Country'] == 'Iraq'])
print(data11.loc[data11['Country'] == 'Switzerland'])
```

```python
# Merging the pre-processed dataset
merged_data_frame = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11],axis=1)
processed_data=merged_data_frame[['Country', 'GDP', 'Population', 'Schooling','Alcohol','BMI','Total_expenditure','thinness__1-19_years','Adult_Mortality','Polio','Diphtheria','Hepatitis_B']]
pre_processed_data=processed_data.iloc[:,10:24] 
print(pre_processed_data)
```

```python
# Replacing the null values in the data dataframe with zero
pre_processed_data.fillna(0,inplace=True)
```

```python
# Re-checking for NULL values
pre_processed_data.isnull().sum()
```

```python
dataframe=original_data[['Country','Year', 'Status', 'Life_expectancy','infant_deaths','percentage_expenditure', 'Measles','under-five_deaths','AIDS','thinness_5-9_years','Income_composition_of_resources']]
print(dataframe)
```

```python
# Merging the two dataframes into single dataset
combined_data = pd.merge(pre_processed_data, dataframe, on='Country')
combined_data
```

```python
# Checking for null values in the final dataset
combined_data.isnull().sum()
```

# Machine Learning Regression algorithms Implementation

## Splitting train & test data

```python
result_data.columns
```

```python
X = result_data[['GDP', 'Population', 'Schooling', 'Alcohol', 'BMI',
       'Total_expenditure', 'thinness_1_to_19_years', 'Adult_Mortality',
       'Polio', 'Diphtheria', 'Hepatitis_B', 'Year', 'Status',
        'infant_deaths', 'percentage_expenditure', 'Measles',
       'under-five_deaths', 'AIDS', 'Region_categorized']]
y = result_data['Life_expectancy']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics

# splits the training and test data set in 80% : 20%
# assign random_state to any value.This ensures consistency.
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=25, stratify=y)

X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression

Linear regression is one of the supervised learning technique. Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x). So, this regression technique finds out a linear relationship between x (input) and y(output). Hence, the name is Linear Regression. If we plot the independent variable (x) on the x-axis and dependent variable (y) on the y-axis, linear regression gives us a straight line that best fits the data points.



```python
# Training the model using linear regression algorithm 
linear_reg = LinearRegression()  
linear_reg.fit(X_train, y_train) 
```

```python
y_pred = linear_reg.predict(X_test)
```

```python
comparision_1 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
comparision_1
```

### Observation
From the above table it is evident that the actual and predicted values are nearly same and there is no much difference in the output.

## Lasso Regression 

The acronym LASSO stands for "Least Absolute Shrinkage and Selection Operator". It is a type of linear regression that uses shrinkage. Shrinkage is where data values are shrunk towards a central point, like the mean. 

```python
# Training the model using lasso regression algorithm 
from sklearn.linear_model import Lasso
lassoRegression = Lasso()
```

```python
lassoRegression.fit(X_train,y_train)
```

```python
y_pred_lasso = lassoRegression.predict(X_test)
```

```python
comparision_2 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred_lasso.flatten()})
comparision_2
```

### Observation
From the above table it is evident that the actual and predicted values are having mixed results in the output with some predicted values are higher than the actual values and vice versa.

# Bias in Artifical Intelligence

Nowadays, Machine learning models are increasingly used to inform high-stakes decisions. Machine learning is always a form of statistical discrimination, the discrimination becomes objectionable when it places certain privileged groups at systematic advantage and certain unprivileged groups at systematic disadvantage. Bias in training data, due to either prejudice in labels or under-/over-sampling, yields models with unwanted bias. To avoid this bias factor, one of the solution was AI Fairness 360.

AI Fairness 360 is an open source toolkit that will help us to examine, report, and mitigate discrimination and bias in machine learning models throughout the AI application lifecycle. Containing over 70 fairness metrics and 10 state-of-the-art bias mitigation algorithms developed by the research community, it is designed to translate algorithmic research from the lab into the actual practice of domains as wide-ranging as finance, human capital management, healthcare, and education. 

The steps involved in the AI Fairness 360 - Demo are as follows:
 - Data 
  - Check 
    - Mitigate
      - Compare

# Pipelining 

Pipelining is used to chain multiple estimators into one. This is done to automate the machine learning process. This is beneficial in data processing. Some codes will transform features such as Normalize numerical values or turn text into vectors, or fill up missing data. They are called transformers. Other codes will predict variables by fitting an algorithm such as a random forest or support vector machine (SVM), they are called estimators. So, in a pipeline, we first sequentially apply a list of transformers (data modelling) and then a final estimator (ML model). In short, pipelines are set up with the fit/transform/predict functionality.

```python
# Importing pipelining from sklearn and calling all the regression functions

from sklearn.pipeline import Pipeline
random_forest = Pipeline(steps=[('regressor_rf', RandomForestRegressor())])
lasso_reg = Pipeline(steps=[('regressor_lasso', Lasso())])
elastic_net = Pipeline(steps=[('regressor_elastic', ElasticNet())])
linear_reg = Pipeline(steps=[('regressor_linear', LinearRegression())])
```

```python
random_forest.fit(X_train, y_train)
```

```python
lasso_reg.fit(X_train, y_train)
```

```python
elastic_net.fit(X_train, y_train)
```

```python
linear_reg.fit(X_train, y_train)
```

```python
# Displaying the regression algorithms scores

score_rf = random_forest.score(X_test, y_test)
print("The pipeline score for Random Forest is", score_rf*100 ,"%")

score_lasso = lasso_reg.score(X_test, y_test)
print("The pipeline score for Lasso Regression is", score_lasso*100 ,"%")

score_elastic = elastic_net.score(X_test, y_test)
print("The pipeline score for Elastic net Regression is", score_elastic*100 ,"%")

score_linear = linear_reg.score(X_test, y_test)
print("The pipeline score for Linear Regression is", score_linear*100 ,"%")
```
